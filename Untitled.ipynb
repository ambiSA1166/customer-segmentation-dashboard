Project Title: Customer Segmentation Engine using RFM Analysis & Machine Learning
Executive Summary
The objective of this project is to transition a business from a generic marketing strategy to a data-driven, personalized approach. By processing transactional data, I built an end-to-end machine learning pipeline that clusters customers into actionable behavioral segments, allowing for optimized marketing budget allocation and targeted retention campaigns.

Methodology & Tech Stack

Data Engineering: Transformed raw, event-level transactional records into customer-level RFM (Recency, Frequency, Monetary) features using Pandas and NumPy.

Data Preprocessing: Addressed extreme right-skewness in monetary values using Log Transformation, and standardized units using StandardScaler to ensure mathematical parity.

Dimensionality Reduction: Applied PCA (Principal Component Analysis) via Scikit-learn to capture ~94% of the dataset's variance within a 2D feature space, removing noise and enabling visualization.

Clustering Algorithm: Utilized K-Means++ to segment the customer base. The optimal number of clusters (k=4) was determined mathematically using the Elbow Method (Inertia analysis).

Data Visualization: Designed heatmaps and scatter plots using Seaborn and Matplotlib to translate mathematical clusters into clear, readable business profiles.

Business Impact & Strategic Recommendations
The model successfully identified four distinct customer cohorts:

Champions (Cluster 2 - High Value): Retain via exclusive VIP programs and early-access rewards.

Loyalists (Cluster 1 - Consistent): Increase lifetime value via targeted cross-selling.

Recent/New (Cluster 0 - Low History): Build habits via onboarding and welcome sequences.

Hibernating (Cluster 3 - Inactive): Minimize ad spend; utilize automated, low-cost win-back emails.

```python
import datetime as dt

# Make sure the date column is in the right format
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

Before this, Python treated the dates like simple sentences (strings). Now that they are datetime objects, Python understands them as actual time. This allows you to perform calculations like "how many days have passed between today and the purchase date."

```python
df_clean['InvoiceDate'].head()
```

#### Recency: When did they last visit the store? (Are they active, or have they abandoned us?)
#### Frequency: How many total trips have they made? (Are they a one-time buyer, or a loyal regular?)
#### Monetary: How much total money have they spent? (Are they a cheap shopper, or a high-value "whale"?)

```python
last_purchase_date = df_clean['InvoiceDate'].max()
```

```python
# We are defining our 'Snapshot Date' as one day after that.
# This gives us a fixed point in time to calculate Recency from.
snapshot_date = last_purchase_date + dt.timedelta(days=1) 
#dt.timedelta(days=1): This is a tool that allows you to add exactly 24 hours (1 day) to a date.
```

now taking every single transaction belonging to Customer X, putting them in a pile, and calculating rfm for that pile.

```python
# We group by the ID, and for each ID, we calculate the three RFM values
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, # Recency: How many days since the last purchase
    'InvoiceNo': 'nunique',                                  # Frequency: How many unique invoices they have
    'TotalSum': 'sum'                                        # Monetary: Total money spent
}).reset_index()
```

```python
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)
```

```python
#Normalize (Squash the outliers)
# Log transform the RFM data
rfm_log = np.log1p(rfm[['Recency', 'Frequency', 'Monetary']])
rfm_log.head()
```

As we saw, your Monetary data has extreme outliers (the "Whale" customers). If you don't "squash" these numbers, your statistics and PCA will be heavily biased by those few individuals.

Action: Apply np.log1p to your Recency, Frequency, and Monetary columns.

Goal: Create a normal-like distribution so the statistical tools work correctly.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)

# Convert back to DataFrame for readability
rfm_scaled_df = pd.DataFrame(rfm_scaled, index=rfm.index, columns=['Recency', 'Frequency', 'Monetary'])
```

The Problem: Recency is in days (1–365), Frequency is in counts (1–200), and Monetary is in dollars (1–280,000). If you don't scale them, Monetary will dominate the math just because the numbers are bigger.

Action: Use StandardScaler to give every column a mean of 0 and a standard deviation of 1.

Goal: Every feature (R, F, and M) now has equal weight in your model.

```python
from sklearn.decomposition import PCA

# Initialize PCA for 2 components
pca = PCA(n_components=2)

# Fit and transform the scaled data
pca_data = pca.fit_transform(rfm_scaled_df)

# Convert to DataFrame for easier plotting
rfm_pca = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'], index=rfm.index)

# Check the variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Information Kept:", sum(pca.explained_variance_ratio_) * 100, "%")
```

```python
rfm_pca.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(rfm_pca['PC1'], rfm_pca['PC2'], alpha=0.5)
plt.title('Customer Segments in 2D Space (PCA)')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Calculate inertia for 1 to 10 clusters
inertia = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(rfm_pca)
    inertia.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, 10), inertia, marker='o', linestyle='--')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Initialize KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
```

```python
# Fit the model to your PCA data (rfm_pca is your 2D data)
cluster_labels = kmeans.fit_predict(rfm_pca)
# Add these labels back into your original RFM table
rfm['Cluster'] = cluster_labels
# Take a look at the first few rows
print(rfm.head())
```

```python
# Group by the cluster and get the average R, F, and M for each
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count' # This shows how many customers are in each cluster
}).rename(columns={'CustomerID': 'Count'})
display(cluster_summary)
```

Using PCA-driven K-Means clustering, the customer base was categorized into four strategic segments:

Cluster 2 (Champions): Top-tier customers with high engagement and spend. Strategy: Reward loyalty with exclusive VIP programs.

Cluster 1 (Loyal Customers): Consistent buyers with steady value. Strategy: Execute upsell and cross-sell initiatives.

Cluster 0 (Recent/New): Recent purchasers in the early journey stage. Strategy: Focus on onboarding and welcome campaigns.

Cluster 3 (Hibernating): Inactive customers with low engagement. Strategy: Deploy automated, low-cost win-back efforts.

Conclusion: This segmentation optimizes marketing ROI by tailoring engagement strategies to the specific behavioral profile of each cluster.
